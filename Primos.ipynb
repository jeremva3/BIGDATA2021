Crear un programa en haskell que genere una lista de nÃºmeros primos siguiendo el siguiente algoritmo:
el elemento n+1 de la lista es primo, si y solo si no es divisible por ninguno del los n elementos anteriores de la lista, caso base la lista del primer elemento con el numero 2.

```python
--NUMEROS PRIMOS
divisible :: Int -> Int -> Bool
divisible x y = x `rem` y == 0
divisores :: Int -> [Int]
divisores x = filter (divisible x) [1..x]
primo :: Int -> Bool
primo x = divisores x == [1,x]
primos_1 :: Int -> [Int]
primos_1 x = filter primo [1..x]
```

```python
--Evaluamos que no sean divisibles x e y
no_divisible :: Int -> Int -> Bool
no_divisible x y =if (x==2 || y==2) then True else x `rem` y /= 0
--Verificamos que el nuevo numero no sea divisible con los --anteriores de la lista
no_divisores :: Int -> [Int]
no_divisores x = [y | y<- primos_1 (x-1), no_divisible x y]

primo_2 :: Int -> [Int]
primo_2 x = if (no_divisores x ==[0]) then [x] else primos_1 (x-1)

primos_2 :: Int -> [Int]
primos_2 n = primo_2 n
```

```python
main = do
  print(primos_2 15)
```
