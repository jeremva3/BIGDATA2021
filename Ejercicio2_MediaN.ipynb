```python
--funcion para determinar el mcm de dos numeros
m1::Int->Int->Int->Int
m1 a b x |a==b=a
     |rem (a*x) b==0=(a*x)
     |otherwise=m1 a b (x+1)
--llamada a la funcion
mcm::Int->Int->Int
mcm a b = m1 a b 1
```

```python
--funcion para determinar el mcm de los n primeros numeros
mcmN :: Int->Int
mcmN n |n==2=2 --si n=2 que devuelva el mcm de(1 y 2)
  --bucle recursivo para calcular el mcm de mas de 3 numeros
       |otherwise=mcm (mcm n (n-1)) (mcmN (n-1))
--funcion para calcular la media 
mediaN ::Int->Float
mediaN n=(fromIntegral (mcmN n))*(fromIntegral(n+1) / 2)
```
